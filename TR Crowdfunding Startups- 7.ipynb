```python
# Translating Categories from Turkish to English
category_list_tr = original_data["category"].unique()  # Categories in Turkish
category_list_tr = ['diğer', 'çevre', 'dans-performans', 'eğitim','film-video-fotoğraf', 'yayıncılık', 'gıda-yeme-içme','kültür-sanat', 
'müzik', 'sağlık-güzellik', 'spor', 'moda', 'teknoloji', 'turizm', 'hayvanlar', 'tasarım', 'sosyal sorumluluk']

# Categories in English
category_list=['other', 'environment', 'dance-performance', 'education','film-video-photography', 'publishing', 'food-eating-drinking',
'culture-art', 'music', 'health-beauty', 'sports', 'fashion','technology', 'tourism', 'animals', 'design', 'social responsibility']

# obtaining a dictionary from the lists
category_dict = dict(zip(category_list_tr, category_list))

original_data['category'] = original_data['category'].map(category_dict) #; original_data["category"].unique()
```

```python
# Translating Gender categories from Turkish to English
project_dict = {
    'erkek': 'male',
    'kadın': 'female',
    'belirsiz': 'unknown'}
original_data['project_owner_gender'] = original_data['project_owner_gender'].map(project_dict) #;original_data['project_owner_gender']
```

```python
# Translating Succuss Status categories from Turkish to English
success_dict= {'başarılı': 'successful', 'başarısız': 'unsuccessful'}
original_data['success_status'] = original_data['success_status'].map(success_dict) #; original_data['success_status']
```

```python
#To sort values it's necessary to convert the dtype in the "support_rate" column from object to integer.

original_data["support_rate"] = (original_data["support_rate"].map(lambda x: x.strip("%"))).astype("int64") #;original_data["support_rate"].info()
```

## Examining the Change in Success Rate by Year

```python
# Successfull Projects for Each Yaer
df_year_success= pd.DataFrame(columns = ["year", "year_project_number", "year_successful_number", "%_year_success_rate"])

for i in range(2013, 2022) :
    year_project_number = original_data[original_data["year"] == i].shape[0]  # row number
    year_successful_number = original_data[(original_data["year"] == i) & (original_data["success_status"] == "successful")].shape[0]
    success_rate = year_successful_number / year_project_number if year_project_number > 0 else 0 # for zero divison error

# new row added to DataFrame df_year_success
    df_year_success.loc[i] = [f"{i}", year_project_number, year_successful_number,int(success_rate*100)]
df_year_success = df_year_success.sort_values(by= "year") ; df_year_success
```

```python
# Successful Projects Numbers for Each Years 
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = np.arange(len(df_year_success))  # number of bars

# placement of bars
ax.bar(index, df_year_success['year_project_number'], bar_width, label='Year Project Number', color='g')
ax.bar(index + bar_width, df_year_success['year_successful_number'], bar_width, label='Year Successful Project Number', color ='b')

ax.set_xlabel('Year')
ax.set_ylabel('Values')
ax.set_title('2013-2021 : Project and Successful Project Numbers')

# Adjusting the places of the years on the x-axis.
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(df_year_success['year'])
ax.legend()
```

```python
# Success Rates for Each Years. Since 2017, success rates seem to be decreasing.

fig, ax = plt.subplots(figsize=(8, 6))
bar_width = 0.45
index = np.arange(len(df_year_success))  # number of bars

print("\nSince 2017, success rates seem to be decreasing.\n")
ax.bar(index + bar_width, df_year_success['%_year_success_rate'], label='Success Rate %', color='darkblue')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(df_year_success['year'])
ax.set_xlabel('Year')
ax.set_ylabel('% Success Rates')
ax.set_title("% Project Success Rate for Each Year")
ax.legend()
```

```python
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
```

## Most Common 50 Words in All Data

```python
#nltk.download('stopwords')

# Getting Turkish stopwords list
stop_words = set(stopwords.words('turkish'))

# Joining all string (word) explanations in 'project_description' columns 
text = " ".join(original_data['project_description'].dropna())

# convert all letters to lowercase and remove punctuation marks.
text = re.sub(r'[^\w\s]', '', text.lower())

# Splitting the words for iteration convenience.
words = text.split()
# Elimination of stop_words
filtered_words = [word for word in words if word not in stop_words]
# Counting word frequency
word_counts = Counter(filtered_words)
# Getting the most common 50 words.
top_50_words = word_counts.most_common(50)

print("\nThe Most Common 50 Words in All Data,\n")
for word, count in top_50_words:
    print(f'{word}: {count}')
```
